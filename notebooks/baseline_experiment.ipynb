# CINIC-10 CNN Baseline Experiment

This notebook implements a baseline convolutional neural network for image classification on the CINIC-10 dataset.

```python
# Setup environment and install dependencies
!pip install -q torch torchvision numpy matplotlib seaborn pandas scikit-learn
```

```python
# Import required libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import torch
import torch.nn as nn
from torchvision import transforms
import sys; sys.path.insert(0, '../src')
from utils import get_device, set_seeds, train_model
from model_architecture import create_baseline_cnn
from data_preprocessing import create_data_generators

set_seeds(42)
device = get_device()
print(f"Using device: {device}")
```

```python
# Define constants
CINIC_CLASSES = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
]

IMG_HEIGHT = 32
IMG_WIDTH = 32
NUM_CLASSES = 10
BATCH_SIZE = 32
```

```python
# Create baseline CNN model
model = create_baseline_cnn().to(device)
print("Baseline CNN model created.")
print(model)
```

```python
# Create data loaders
DATA_DIR = os.path.join('..', 'data')
TRAIN_DIR = os.path.join(DATA_DIR, 'train')
VAL_DIR = os.path.join(DATA_DIR, 'valid')

train_loader, val_loader = create_data_generators(
    TRAIN_DIR, VAL_DIR, batch_size=BATCH_SIZE
)
print(f"Train loader: {len(train_loader)} batches")
print(f"Val loader:   {len(val_loader)} batches")
```

```python
# Create and display the baseline model architecture
print("Baseline CNN model architecture:")
print(model)
```

```python
# Train the baseline model
print("Starting training...")
epochs = 5

optimizer = torch.optim.Adam(model.parameters())
history = train_model(model, train_loader, val_loader, optimizer,
                      epochs=epochs, device=device)

print("\nTraining completed.")
for i, (tl, ta, vl, va) in enumerate(
    zip(history['train_loss'], history['train_acc'],
        history['val_loss'],   history['val_acc']), 1
):
    print(f"Epoch {i:2d}: Train Loss: {tl:.3f}, Train Acc: {ta:.3f}, "
          f"Val Loss: {vl:.3f}, Val Acc: {va:.3f}")
```

```python
# Visualize training results
epochs_range = range(1, len(history['train_acc']) + 1)
train_accs   = history['train_acc']
val_accs     = history['val_acc']
train_losses = history['train_loss']
val_losses   = history['val_loss']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Baseline CNN Training Results', fontsize=16, fontweight='bold')

# Accuracy
axes[0].plot(epochs_range, train_accs,   'o-', label='Training Accuracy')
axes[0].plot(epochs_range, val_accs,     's-', label='Validation Accuracy')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Training vs Validation Accuracy')
axes[0].legend()
axes[0].grid(True)

# Loss
axes[1].plot(epochs_range, train_losses, 'o-', label='Training Loss')
axes[1].plot(epochs_range, val_losses,   's-', label='Validation Loss')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Loss')
axes[1].set_title('Training vs Validation Loss')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
